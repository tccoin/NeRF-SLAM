```python
import cv2
import numpy as np
import glob
import matplotlib.pyplot as plt

def image_resize(image, width = None, height = None, inter = cv2.INTER_AREA):
    # initialize the dimensions of the image to be resized and
    # grab the image size
    dim = None
    (h, w) = image.shape[:2]

    # if both the width and height are None, then return the
    # original image
    if width is None and height is None:
        return image

    # check to see if the width is None
    if width is None:
        # calculate the ratio of the height and construct the
        # dimensions
        r = height / float(h)
        dim = (int(w * r), height)

    # otherwise, the height is None
    else:
        # calculate the ratio of the width and construct the
        # dimensions
        r = width / float(w)
        dim = (width, int(h * r))

    # resize the image
    resized = cv2.resize(image, dim, interpolation = inter)

    # return the resized image
    return resized


dataset_dir = '/home/junzhe/dataset/nerf-cube-diorama-dataset/room_underwater_ugan/images'
image_paths = sorted(glob.glob(f'{dataset_dir}/*.png'))

w,h = 640,480

for i, image_path in enumerate(image_paths):
    if not image_path.endswith('.depth.png'):
        w_, h_ = cv2.imread(image_path).shape[:2]
        img = cv2.imread(image_path)
        # resize but keep the aspect ratio, and then crop the image to w,h
        if w_/w > h_/h:
            img = image_resize(img, width=w)
            img = img[(img.shape[0]-h)//2:(img.shape[0]-h)//2+h, :]
        else:
            img = image_resize(img, height=h)
            img = img[:, (img.shape[1]-w)//2:(img.shape[1]-w)//2+w]
        print(img.shape)
        cv2.imwrite(image_path, img)
        # plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))


```
